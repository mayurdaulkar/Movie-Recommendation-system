```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv("file1.csv")
```

```python
df.head()
```

```python
df["genres"] = df["genres"].str.replace(r"[^\w\s\,\[\]]" , "" , regex = True)
df["keywords"] = df["keywords"].str.replace(r"[^\w\s\,\[\]]", "" , regex = True)
df["cast"] = df["cast"].str.replace(r"[^\w\s\,\[\]]" , "" , regex = True)
df["crew"] = df["crew"].str.replace(r"[^\w\s\,\[\]]" , "" , regex = True)

```

```python
df["overview"] = df["overview"].apply(lambda x: x.split())
```

```python
df.head()
```

```python
df["cast"] = df["cast"].str.replace(" " , "")
df["crew"] = df["crew"].str.replace(" " , "")
df["genres"] = df["genres"].str.replace(" " , "")
df["keywords"] = df["keywords"].str.replace(" " , "")
```

```python
df.head()
```

```python
df["overview"] = df["overview"].apply(lambda x:[x])
df["cast"] = df["cast"].apply(lambda x: [x])
df["crew"] = df["crew"].apply(lambda x: [x])
df["keywords"] = df["keywords"].apply(lambda x: [x])
df["genres"] = df["genres"].apply(lambda x: [x])
```

```python
df["tags"] = df["overview"] +df["genres"]+df["keywords"]+df["cast"]+df["crew"]
```

```python
df.head()
```

```python
df.drop(columns = ["genres" , "overview" , "keywords" , "cast"  , "crew"] , inplace = True)
```

```python
df.to_csv("file2.csv" , index = False)
```
