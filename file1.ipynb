```python
df1 = pd.read_csv("credits.csv")
```

```python
movie = df.merge(df1 , on = "title")
```

```python
movie.head()
```

```python
movie.columns
```

```python
movie = movie[["genres" , "id" ,"keywords" , "title" , "overview" , "cast" ,"crew"]]
```

```python
movie.isnull().sum()
```

```python
movie = movie.dropna()
```

```python
movie.isnull().sum()
```

```python
import ast
```

```python
movie["genres"]  = movie["genres"].apply(ast.literal_eval)
```

```python
movie["genres"] = movie["genres"].apply(lambda x:[item["name"] for item in x])
```

```python
movie.head()
```

```python
movie["keywords"] = movie["keywords"].apply(ast.literal_eval)
```

```python
movie["keywords"] = movie["keywords"].apply(lambda x:[item["name"] for item in x])
```

```python
movie.head()
```

```python
movie["cast"] = movie["cast"].apply(ast.literal_eval)
```

```python
movie["cast"] = movie["cast"].apply(lambda x:[item["name"]for item in x][:3])
# movie.iloc[0]["cast"]
```

```python
movie.head()
```

```python
movie["crew"] = movie["crew"].apply(ast.literal_eval)
```
